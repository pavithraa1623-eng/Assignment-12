Additional Notes:
1.Explain Bagging and Boosting methods. How is it different from each other.

Bagging:

Bagging (Bootstrap Aggregating) is an ensemble learning technique used to reduce variance and avoid overfitting.

It trains multiple models independently on different random samples (with replacement) of the dataset and combines their predictions using averaging (regression) or majority voting (classification). 

A common example of bagging is Random Forest.

Boosting:

Boosting is an ensemble method that mainly reduces bias and improves accuracy.

 Models are trained sequentially, and each new model focuses more on correcting the errors made by the previous ones by giving higher weight to misclassified data points.

	 Popular boosting algorithms include AdaBoost, Gradient Boosting, and XGBoost.

Difference:

		Bagging trains models in parallel and treats all data points equally, while Boosting trains models sequentially and gives more importance to wrongly predicted samples.

2. Explain how to handle imbalance in the data.

Imbalanced data occurs when one class has significantly more samples than the other(s), which can cause a model to be biased toward the majority class. The following techniques are commonly used to handle class imbalance:

1. Data-Level Methods

These methods modify the training data to balance class distribution.

Undersampling: Reduces the number of majority-class samples.

Risk: Loss of useful information.

Oversampling: Increases minority-class samples by duplication.

Example: Random Oversampling.

Synthetic Data Generation: Creates new minority-class samples artificially.
Examples: SMOTE, ADASYN.

2. Algorithm-Level Methods

These methods modify the learning algorithm rather than the data.

Class Weighting: Assigns higher weights to minority-class samples so misclassification is penalized more.

Cost-Sensitive Learning: Different misclassification costs are defined for different classes.

3. Ensemble Methods

Balanced Bagging: Combines undersampling with bagging.

Boosting Methods: Algorithms like AdaBoost or Gradient Boosting can focus more on misclassified (often minority-class) samples.

4. Appropriate Evaluation Metrics

Accuracy is misleading for imbalanced data. Use:
